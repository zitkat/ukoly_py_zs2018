```python
import pandas as pd
```

```python
napoje_df = pd.read_excel("lb_data_clean.xls")
# napoje_df = pd.read_excel("lb_data_clean.xls", dtype={'ID': object})
napoje_df
```

```python
kod2typ = pd.read_excel("kod2typ.xls")
kod2typ
```

```python
for index, pol in napoje_df.iterrows():
    if pd.isna(pol["Kód"]):
        continue
    napoje_df.at[index, "Reference"] = (
                        pol["Kód"] + 
                        str(pol["Číslo"]) + 
                        kod2typ[kod2typ["Kód"].str.contains(pol["Kód"])]["TYP"].values[0])
```

```python
napoje_df
```

```python
napoje_df["Číslo"] = napoje_df["Číslo"].apply('000{}'.format)
```

```python
napoje_df
```

```python
napoje_df.to_excel("ld_results_fromclean.xls")
```
